```python
dados_pessoais = []
produtos_eletronicos = ["Computador Dell", "Video Game - PS5", " Máquina de Lavar", "Tv 100 Polegadas"]
preco_produtos = [5500.00, 6500.00, 1200.00, 12000.00]
carrinho = []
meus_valores = []
forma_pagamento = ["Cartão de Crédito", "Pix", "Boleto"]
tipo_pagamento = []

nome = str(input("Digite seu nome"))
print(f"{nome} Seja Bem Vindo a BlueStore Eletronicos SA !!!!")
print(f"{nome} Aceitamos como Forma de Pagamento Cartões de Crédito: Visa, Elo e Amex!!")
print(f"{nome} Agora vamos realizar o cadastro dos seu dados:")

nome_completo = input("Digite seu nome completo:")
data_de_nascimento  = input("Digite sua data de nascimento:")
endereco = input ("Digite seu endereço:")
cidade = input ("Digite sua Cidade")
cpf = input(" Digite seu CPF")

dados_pessoais.append({ 
    "nome_completo": nome_completo, 
    "data_de_nascimento": data_de_nascimento, 
    "endereco": endereco, 
    "cidade": cidade, 
    "cpf": cpf
})

print("""
""")

print("Dados pessoais cadastrados:") 
for pessoa in dados_pessoais: 
    print(f"""
          Nome Completo: {pessoa['nome_completo']}, 
          Data de Nascimento: {pessoa['data_de_nascimento']}, 
          Endereço: {pessoa['endereco']}, 
          Cidade: {pessoa['cidade']}, 
          CPF: {pessoa['cpf']}
          """)

print (f"{nome}, Agora que você já realizou seu cadastro voce vai selecionar o produto que deseja comprar")

produto1 = int(input(f"""Digite  o código do produto que você deseja adquirir
0 - Computador Dell
1 - Video Game - PS5 
2 - Máquina de Lavar
3 - Tv 100 Polegadas                     
"""))

produto2 = int(input(f"""Digite  o código do produto que você deseja adquirir
0 - Computador Dell
1 - Video Game - PS5 
2 - Máquina de Lavar
3 - Tv 100 Polegadas                     
"""))

carrinho = [produtos_eletronicos[produto1], produtos_eletronicos[produto2]]
meus_valores = [preco_produtos[produto1], preco_produtos[produto2]]

soma = sum(meus_valores)
desconto = soma *0.10
desconto_final = soma - desconto

pagamento = int(input(f""" {nome} seus produtos já estão no cesto de pamento agora selecione a forma de pagamento
0 - Cartão de Crédito
1 - Pix
2 - Boleto
"""))
print(f"""
============================================================ Nota Fiscal ================================================================================
{produtos_eletronicos[produto1]} R$ {preco_produtos[produto1]:.2f}
{produtos_eletronicos[produto2]} R$ {preco_produtos[produto2]:.2f}

======================================================= Valor Total da Compra ===========================================================================
O valor total da compra dos produtos acima foram de R$ {soma:.2f}             
""")
print("============================================ Forma de Pagamento =========================================================================================")
if pagamento == 0:
    parcelas = soma / 12
    print(f"{nome}, o valor total da compra é de R$ {soma:.2f} podera ser parcelado em até 12 vezes sem juros no Cartão de Crédito!!! Sendo 12 parcelas de R$ {parcelas:.2f} ")
elif pagamento == 1:
    #desconto = soma - (soma * 0.10)
    print(f" {nome}, como o pagamento será via Pix você terá 10% de desconto em cima do valor total de {desconto:.2f}, o valor final da compra será de R$ {desconto_final:.2f}")
else:
    print (f"{nome}. Infelizmente para o pagamento de Boleto não temos desconto,  o valor total da compra é de {soma:.2f} deverá ser pago á Vista")


print(f"{nome}, seu pagamento foi realizado com sucesso!!!! A BlueStore Eletronicos SA agradece a preferência e volte sempre!!!!")

```
